```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Conv2D,Flatten
from keras.models import Sequential
from keras.datasets import mnist
from keras.utils import to_categorical
from tensorflow.keras import layers
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train/255
x_test=x_test/255
```

```python
x_train.shape
# if your data is in shape
# (batch, height, width)
# you must reshape it into  (batch, height, width, channels) for CNN
```

```python
x_train=x_train.reshape(len(x_train),28,28,1)
x_test=x_test.reshape(len(x_test),28,28,1)
```

```python
model=keras.Sequential([
    Conv2D(filters=32,kernel_size=(3,3),padding='valid',activation='relu',input_shape=((28,28,1))),
    Conv2D(filters=32,kernel_size=(3,3),padding='valid',activation='relu'),
    Conv2D(filters=32,kernel_size=(3,3),padding='valid',activation='relu'),
    ##2d to 1d
    layers.Flatten(),

    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile (
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs=5,validation_split=0.2)
```

```python
y_pred=model.predict(x_test)
```

```python
y_test[0]
```

```python
import numpy as np
np.argmax(y_pred[0])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,np.argmax(y_pred,axis=1)))
```

```python
##WITH PADDING

model2=keras.Sequential([
    Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu',input_shape=((28,28,1))),
    Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu'),
    Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu'),
    ##2d to 1d
    layers.Flatten(),

    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])
```

```python
model2.summary()
```

```python
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model2.fit(x_train,y_train,epochs=5,validation_split=0.2)
```

```python
y_pred2=model2.predict(x_test)
```

```python
y_pred2=np.argmax(y_pred2,axis=1)
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from keras.src.layers.pooling.max_pooling2d import MaxPooling2D
##with max pooling
model3= keras.Sequential([
    Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu',input_shape=((28,28,1))),
    MaxPooling2D(pool_size=(2,3)),
    Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu'),
    MaxPooling2D(pool_size=(2,3)),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])
```
