```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv('sonar_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().isnull()
```

```python
df.columns
```

```python
df['R'].value_counts()
```

```python
df1=df
```

```python
df1=df.replace({'R': {'R':1,'M':0}})
```

```python
df1['R']
```

```python
x=df1.drop('R',axis=1)
y=df1['R']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
x_train.shape,x_test.shape
```

```python
modeld=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
modeld.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

modeld.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
y_train.shape
```

```python
modeld.evaluate(x_test,y_test)
```

```python
y_predict=modeld.predict(x_test)
y_predict

##round off the values
y_predict=np.round(y_predict)
y_predict
```

```python
y_test
```

```python
y_predict[0
          ]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_predict)) ## y_true=y_test , y_pred=y_lables
```

```python
##introducing dropout layers
model=keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
y_pred=np.round(y_pred)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```
