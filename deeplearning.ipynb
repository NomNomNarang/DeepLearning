```python
import tensorflow as tf
from tensorflow import keras
```

```python
## handwritten digits
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#0->black 255->white
```

```python
(x_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
len(x_train)
```

```python
len(X_test)
```

```python
##scaling increases the accuracy
x_train=x_train/255
X_test=X_test/255
```

```python
y_train[90]
```

```python
##flatten our dataset->
x_train_flatten=x_train.reshape(len(x_train),28*28)
x_test_flatten=X_test.reshape(len(X_test),28*28)
```

```python
x_train_flatten[0]
```

```python
##creating a simple neural network
model=keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam',  ##how model will learn
    loss='sparse_categorical_crossentropy', #how wrong guesses it will make
    metrics=['accuracy'] #what you want to keep a track of while model learns
)

model.fit(x_train_flatten,y_train,epochs=5) ##training actually happens

##KERAS SEQUENTIAL AS STACK OF LAYERS
#-->IN NEURAL NETWORK. DATA FLOW LAYER BY LAYER

#keras layers dense([...])
#-->dense means fully connected layer, every input neuron
#is connected to every neuron in this layer

##OUTPUT
#--->10 neurons each neuron will give one output value

##input_shape=(784,)  ONLY REQUIRED IN THE FIRST LAYER OF NEURAL NETWORK
#-->this tells keras how many inputs each neuron will receive.
#--> it is reshaped so each image is a list of 784 numbers

##activation function
#-->this is the acivaton functions of neurons,
#activation function is of different types== SIGMOID,RELU,SOFTMAX etc
```

```python
model.evaluate(x_test_flatten,y_test)
```

```python
X_test[0]
```

```python
plt.matshow(X_test[1])
```

```python
y_predicted=model.predict(x_test_flatten)
y_predicted[1]
```

```python
np.argmax(y_predicted[1])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
import seaborn as sns
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('actual')
```

```python
#gradient descent
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
age=np.array([22,25,47,52,46])
affordibility=np.array([1,0,1,0,1])
bought_insurance=np.array([0,0,1,0,1])
df=pd.DataFrame({'age':age,'affordibility':affordibility,'bought_insurance':bought_insurance})
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['age','affordibility']], df.bought_insurance,test_size=0.2,random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
#scaling the data
x_train['age']=x_train['age']/100
x_test['age']=x_test['age']/100
x_train
```

```python
x_test
```

```python
model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train,y_train,epochs=500)
```

```python
model.evaluate(x_test,y_test)
```

```python
model.predict(x_test)  #predicted it correctly
```

```python
coef,intercept=model.get_weights()
coef,intercept   ##first two gives w1,w2 and second ones from b
```

```python
#3if i want to see how weight performs

new_weights = np.array([[0.56999964], [0.62993133]])  # shape (2,1)
new_bias = np.array([-0.44537964])                     # shape (1,)

model.set_weights([new_weights, new_bias])
model.evaluate(x_test, y_test)
model.predict(x_test)

```

```python
coef,inter=model.get_weights()
coef,inter
```

```python
##if i wnat my model to learn further with these new weights
model.set_weights([new_weights, new_bias])
model.fit(x_train, y_train, epochs=500)
```
